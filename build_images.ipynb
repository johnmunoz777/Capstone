### Total Frames in a 1-Minute Video with 60 FPS

If your video is **1 minute** long and has **60 frames per second (FPS)**, you can calculate the total number of frames in the video using the following formula:

- **Video Duration** = 1 minute = 60 seconds  
- **Frames per second (FPS)** = 60

#### Formula:
$$
\text{Total frames} = \text{FPS} \times \text{Video Duration (in seconds)}
$$
$$
\text{Total frames} = 60 \times 60 = 3600
$$

So, the total number of frames in the video is **3600 frames**.

---

### Selecting Every 5th Frame

If you want to select every 5th frame from the video, you can calculate the total number of images you will export by dividing the total number of frames by 5.

#### Formula:
$$
\text{Number of images} = \frac{\text{Total frames}}{5}
$$
$$
\text{Number of images} = \frac{3600}{5} = 720
$$

Thus, if you take every 5th frame, you will get **720 images**.

---

### How to Extract 20 Images from the Video

If you want to extract exactly **20 images** from the video, you need to calculate how many frames you need to skip between each image.

#### Formula to Calculate the Interval:
$$
x = \frac{\text{Total frames}}{\text{Desired number of images}} = \frac{3600}{20} = 180
$$

So, to get **20 images**, you will need to extract every **180th frame** from the video.

```python
import cv2
import os

def export_frames(video_path, total_images, output_folder='john'):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the total number of frames in the video
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Calculate the interval to capture frames (frame_interval)
    frame_interval = total_frames // total_images

    # Loop through the video to capture the required frames
    for i in range(total_images):
        # Set the video capture to the desired frame (every frame_interval-th frame)
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * frame_interval)

        # Read the frame
        ret, frame = cap.read()

        # Check if the frame was successfully read
        if ret:
            # Resize the frame to (416, 416)
            resized_frame = cv2.resize(frame, (416, 416))

            # Define the filename for the frame (e.g., john_1.jpg, john_2.jpg, ...)
            filename = os.path.join(output_folder, f'john_{i + 1}.jpg')

            # Save the frame as a .jpg image
            cv2.imwrite(filename, resized_frame)

    # Release the video capture object
    cap.release()

    print(f"Exported {total_images} frames to {output_folder}.")

# Example usage:
video_path = 'john.mp4'  # Replace with your video file path
total_images = 20  # You can change this value to get more or fewer images

# Call the function
export_frames(video_path, total_images)
```

```python
import cv2
import os

# Function to detect a face and output YOLO format coordinates
def detect_face_and_generate_yolo_format(image_path, output_txt_path, class_id=0):
    # Load the pre-trained Haar Cascade for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    # Read the input image
    img = cv2.imread(image_path)
    # Convert image to grayscale (required for face detection)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # If faces are detected
    if len(faces) == 0:
        print("No face detected!")
        return

    # Get the first face detected (can be adjusted to handle multiple faces)
    x, y, w, h = faces[0]

    # Get the image dimensions
    image_height, image_width, _ = img.shape

    # Convert the bounding box coordinates to YOLO format (normalized)
    x_center = (x + w / 2) / image_width
    y_center = (y + h / 2) / image_height
    width = w / image_width
    height = h / image_height

    # Prepare the YOLO formatted string
    yolo_format = f"{class_id} {x_center} {y_center} {width} {height}"

    # Write the YOLO format to a .txt file
    with open(output_txt_path, 'w') as file:
        file.write(yolo_format)

    print(f"YOLO format bounding box saved to {output_txt_path}")

    # Optionally, draw the bounding box on the image and show it (for visual verification)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img, "Face Detected", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image with the bounding box using Matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis for a cleaner view
    plt.show()
    # Save the image with the bounding box for visual reference (optional)
    cv2.imwrite('detected_face.jpg', img)

# Example usage:
image_path = 'john/john_9.jpg'  # Path to the input image (the face image)
output_txt_path = 'johns_image_test_yolo.txt'  # Path to save the YOLO annotation file

# Call the function to detect face and save YOLO annotations
detect_face_and_generate_yolo_format(image_path, output_txt_path)
```

```python
import cv2
import matplotlib.pyplot as plt
import csv

# Define class labels
class_labels = {
    0: "John",
    1: "ten",
    2: "twenty",
    3: "fifty",
    4: "hundred"
}

def draw_bounding_boxes(image_path, annotation_file_path):
    # Open the file with YOLO annotations
    with open(annotation_file_path, "r") as fs:
        fs = csv.reader(fs, delimiter=" ")
        annotations = list(fs)

    # Read the image
    image = cv2.imread(image_path, 1)
    image_height, image_width, _ = image.shape

    # Loop over each row in the annotation file
    for row in annotations:
        # Parse the YOLO-style annotation
        class_id = int(row[0])  # Class ID (you can use this for debugging or filtering)
        normalized_x_center = float(row[1])  # Normalized x center
        normalized_y_center = float(row[2])  # Normalized y center
        normalized_width = float(row[3])  # Normalized width
        normalized_height = float(row[4])  # Normalized height

        # Reverse normalization to get pixel values
        x_center = normalized_x_center * image_width
        y_center = normalized_y_center * image_height
        box_width = normalized_width * image_width
        box_height = normalized_height * image_height

        # Convert center coordinates to top-left corner coordinates
        top_left_x = int(x_center - (box_width / 2))
        top_left_y = int(y_center - (box_height / 2))

        # Calculate bottom-right corner coordinates
        bottom_right_x = top_left_x + int(box_width)
        bottom_right_y = top_left_y + int(box_height)

        # Draw the bounding box on the image (Green color, thickness 2)
        image = cv2.rectangle(image, (top_left_x, top_left_y), (bottom_right_x, bottom_right_y), (0, 255, 0), 2)

        # Get the class label
        label = class_labels.get(class_id, "Unknown")  # Default to "Unknown" if class_id is not in dictionary

        # Add the class label text above the bounding box (Red text)
        cv2.putText(image, label, (top_left_x, top_left_y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

    # Convert BGR image to RGB for displaying in Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the final image with all bounding boxes using Matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis for a cleaner view
    plt.show()


# Example usage:
image_path = 'john/john_9.jpg'  # Path to the input image (the face image)
output_txt_path = 'johns_image_test_yolo.txt'  # Path to save the YOLO annotation file

# Example usage of the function
draw_bounding_boxes(image_path, output_txt_path )
```
